````python
import json
import pickle
import tweepy as tw
import pandas as pd
````

We stores the api keys in ```twitter_secret```.

````python
TWITTER_SECRET = '../twitter_secret'

with open(TWITTER_SECRET, 'r') as f:
    keys = json.load(f)
````

````python
auth = tw.OAuthHandler(keys['consumer_key'], keys['consumer_secret'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tw.API(auth, wait_on_rate_limit=True)
````

````python
# Copied Amy's code here..
# See available regions
# api.trends_available() # worldwide has woeid of 1

WOEID = 23424977 # united states

# Get worldwide trending topics and number of Tweets for each
trends1 = api.trends_place(WOEID)
trends = [(trend['name'], trend['tweet_volume']) for trend in trends1[0]['trends']]

# Create pandas df
trends_df = pd.DataFrame(trends, columns=['Topic', 'Count'])
trends_df.head(50)
````

## search topics

````python
# Define the search term and the date_since date as variables
search_words = "#DonateToAndrewYang"
date_since = "2019-12-01"
````

````python
# Collect tweets
tweets = tw.Cursor(api.search,
              q=search_words,
              lang="en",
              since=date_since).items(1000)
tweets
````

````python
tweet_list = [tweet for tweet in tweets]
````

````python
for tw_ in tweet_list[:10]:
    print(tw_._json)
````

````python
def gen_tweets_filename(search_words):
    fname = 'tweets_'
    if '#' == search_words[0]:
        return fname + search_words[1:] + '.pkl'
    else:
        return fname + search_words + '.pkl'
    
output_filename = gen_tweets_filename(search_words)
print("# pickle the list of tweets into {}".format(output_filename))
with open(output_filename, 'wb') as f:
    pickle.dump(tweet_list, f)
````
